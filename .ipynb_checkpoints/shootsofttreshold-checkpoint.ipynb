```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img2 = cv2.imread('shootanalysisdataset/schotanalyse/20200923_131212.jpg')
kernel = np.ones((10, 10), np.uint8)

#img = cv2.erode(img2, kernel, iterations=1)
img = cv2.dilate(img2, kernel, iterations=4)
img3 = cv2.erode(img2, kernel, iterations=3)

imagem = cv2.bitwise_not(img3)
```

```python
plt.imshow(imagem, cmap='gray')
```

```python
ret,thresh1 = cv2.threshold(imagem,127,255,cv2.THRESH_TOZERO_INV)
```

```python
plt.imshow(thresh1)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
ret,th1 = cv2.threshold(img,20,50,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
ret,th1 = cv2.threshold(img,20,200,cv2.THRESH_BINARY)
show_pic(th1)
```
