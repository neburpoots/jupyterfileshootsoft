```python
import math

contour_list = []
center_coordinates = np.zeros((0,3))
area_list = np.zeros((0,2))
average_list = np.zeros((0,2))


def scale_contour(cnt, scale):
    M = cv2.moments(cnt)
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])

    cnt_norm = cnt - [cx, cy]
    cnt_scaled = cnt_norm * scale
    cnt_scaled = cnt_scaled + [cx, cy]
    cnt_scaled = cnt_scaled.astype(np.int32)

    return cnt_scaled


def calc_average(contour):
    M = cv2.moments(contour)
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])
        
    avgTotal = 0
    count = 0
    for x, point in enumerate(contour):
        dif_x = cx - point[0][0]
        dif_y = cy - point[0][1]

        dif_x = abs(dif_x)
        dif_y = abs(dif_y)

        distance = math.sqrt((dif_x + dif_y))

        avgTotal += distance
        count = count + 1
            
    
    return avgTotal / count

for x, contour in enumerate(contours):
    approx = cv2.approxPolyDP(contour,0.01*cv2.arcLength(contour,True),True)
    area = cv2.contourArea(contour)
    if ((len(approx) > 8) & (area > 30000) ):
        contour_list.append([x, contour])
        area_list = np.vstack([area_list, [x, area]])

        M = cv2.moments(contour)
        #GETS THE CENTER COORDINATES PLAN IS TO REMOVE OUTLIERS
        cx = int(M['m10']/M['m00'])
        cy = int(M['m01']/M['m00'])
        center_coordinates = np.vstack([center_coordinates, [cx, cy, x]])

x_median = np.median(center_coordinates[:,0])
y_median = np.median(center_coordinates[:,1])

#print(x_median)
#print(len(contour_list))

#removes contour from contour list if it is not close enough to the medium this is done to filter out contours
for coordinate in center_coordinates:

    if(((x_median * 1.2) < coordinate[0] or coordinate[0] < (x_median * 0.8)) or 
       ((y_median * 1.2) < coordinate[1] or coordinate[1] < (y_median * 0.8))):

        array_of_keys = np.array([inner_arr[0] for inner_arr in contour_list])
        indices = np.where(array_of_keys == int(coordinate[2]))
        
        mask = center_coordinates[:, 2] != coordinate[2]
        center_coordinates = center_coordinates[mask]
        
        contour_list.pop(indices[0][0])
        
        #print(indices[0][0])
        #print(coordinate[2])
        #print(coordinate[0])
    

# #calculates average distance between middle point and contour edge
for contour in contour_list:
    M = cv2.moments(contour[1])
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])
    cv2.circle(crop_img, (cx, cy), 7, (255, 0, 255), -1)
    
    avgTotal = 0
    count = 0
    for x, point in enumerate(contour[1]):
        dif_x = cx - point[0][0]
        dif_y = cy - point[0][1]
        
        dif_x = abs(dif_x)
        dif_y = abs(dif_y)
        
        distance = math.sqrt((dif_x + dif_y))
        
        avgTotal += distance
        count = count + 1
        
    
    average_list = np.vstack([average_list, [(avgTotal / count), contour[0]]])

def find_inner_array(arr, value):
    for inner_arr in arr:
        if inner_arr[1] == value:
            return inner_arr
    return None
    
counter = 0
while counter < (len(contour_list) - 1):
    checkLength = len(contour_list)
    contour = contour_list[counter]
    M = cv2.moments(contour[1])
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])
    average = find_inner_array(average_list, contour[0])
    min_average = average[0] * 0.7
    max_average = average[0] * 1.3
    for x, point in enumerate(contour[1]):
        dif_x = cx - point[0][0]
        dif_y = cy - point[0][1]
        
        dif_x = abs(dif_x)
        dif_y = abs(dif_y)
        
        distance = math.sqrt((dif_x + dif_y))
        if min_average > distance or max_average < distance:
            print(counter)
            print(average[0])
            print(distance)
            
            contour_list.pop(counter)
            break
    
    if checkLength is len(contour_list):
        counter = counter + 1
            
            
        
            


area_list = area_list[area_list[:, 1].argsort()]


#ALGORITHM TO REMOVE DOUBLE DETECTIONS
#
#
#
#
#

for x, area in enumerate(area_list):
    #print(area[1])
    if x < (len(area_list) - 1):
        next_area = area_list[x + 1][1]

        if (area[1] * 1.15) > next_area:
            next_key = area_list[x + 1][0]
            current_key = int(area[0])
            
            array_of_keys = np.array([inner_arr[0] for inner_arr in contour_list])
            indices = np.where(array_of_keys == int(current_key))
            if len(indices[0]) > 0:
                #print(indices[0][0])
                contour_list.pop(indices[0][0])
    

            
            
contours_final = [subarr[1] for subarr in contour_list]

contours_final_sorted = sorted(contours_final, key=calc_average)


final_area_list = np.zeros((0,2))

double_contour_area = cv2.contourArea(black_contour)
area_min = double_contour_area * 0.7
area_max = double_contour_area * 1.3
base_contour = 0
print(double_contour_area)
print("")

for x, contour in enumerate(contours_final_sorted):
    area = cv2.contourArea(contour)
    print(area)
    
    final_area_list = np.vstack([final_area_list, [x, area]])

    if area_min < area and area_max > area:
        base_contour = contour






generate_counter = 0




while generate_counter < 10:
    
    lastarea = cv2.contourArea(contours_final_sorted[len(contours_final_sorted) - 1])
    secondtoLast = cv2.contourArea(contours_final_sorted[len(contours_final_sorted) - 2])

    if generate_counter < (len(contours_final_sorted) - 1):     

        contour = contours_final_sorted[generate_counter]
        append = False
        area = cv2.contourArea(contour)
        average = calc_average(contour)
        
        next_average = calc_average(contours_final_sorted[generate_counter + 1])
        print("area", area)
        print(average)
        print(next_average)
#         if (next_average - average) > 10 and generate_counter == 1:
#             append = True
            
#             new_average = average + 4
#             extra_option = 10 - generate_counter
#             scale = (new_average * (1 + (extra_option / 10))) / average 
#             print("scale", scale)
#             contours_final_sorted.insert(generate_counter + 1, scale_contour(contour, scale))
        if (next_average - average) < 2:
            contours_final_sorted.pop(generate_counter)
            generate_counter = generate_counter - 1
#     else:
#         #CHECKS IF ALL RINGS ARE PRESENT IF NOT GENERATE THEM
#         if len(contours_final_sorted) < 10:
#             length = len(contours_final_sorted)
# #             append = False
# #             area = cv2.contourArea(contour)

# #             average = calc_average(contour)
            
            
#             print("length", len(contours_final_sorted))
            
#             print("average", average)
#             new_average = average + 10 + (20 - (generate_counter * 3))
#             scale = new_average / average 
#             print("scale", scale)
#             if length == 7:
#                 contours_final_sorted.append(scale_contour(black_contour, 2.3))
#             elif length == 6:
#                 contours_final_sorted.append(scale_contour(black_contour, 2.1))
#             elif length == 8:
#                 contours_final_sorted.append(scale_contour(black_contour, 2.5))
#             elif length == 9:
#                 contours_final_sorted.append(scale_contour(black_contour, 2.7))
        
    # this does a check to make sure the last circle is not the edge of the paper
     
    
    generate_counter = generate_counter + 1
    
    
    
    if generate_counter > 9:
        break


cv2.drawContours(crop_img, contours_final_sorted,  -1, (255,0,0), 2)


#OM DE CIRKEL NOG BETER TE DETECTEREN PROBEER HET VERSCHILLENDE BLUR LEVELS
#IDEE ALS NIET ALLE CIRKELS GEVONDEN WORDEN PROBEREN MET ANDERE BLUR


#double_contour = scale_contour(contour_list[0], 0.45)

#cv2.drawContours(img2, [double_contour],  -1, (255,0,0), 2)

display(crop_img)
```

```python
import random
import csv
import os


def overlay_transparent(background, overlay, x, y):

    background_width = background.shape[1]
    background_height = background.shape[0]

    if x >= background_width or y >= background_height:
        return background

    h, w = overlay.shape[0], overlay.shape[1]

    if x + w > background_width:
        w = background_width - x
        overlay = overlay[:, :w]

    if y + h > background_height:
        h = background_height - y
        overlay = overlay[:h]

    if overlay.shape[2] < 4:
        overlay = np.concatenate(
            [
                overlay,
                np.ones((overlay.shape[0], overlay.shape[1], 1), dtype = overlay.dtype) * 255
            ],
            axis = 2,
        )

    overlay_image = overlay[..., :3]
    mask = overlay[..., 3:] / 255.0

    background[y:y+h, x:x+w] = (1.0 - mask) * background[y:y+h, x:x+w] + mask * overlay_image

    return background

def decide_shot_img(points, directory_size_white, directory_size_black):
    
    #standard white shot
    random_shot = random.randint(1, directory_size_white)
    shot = cv2.imread(f'shootanalysisdataset/schotennobg/shotwhite/{random_shot}.png', cv2.IMREAD_UNCHANGED)
    
    #startpoint is in black contour select different shot
    for point in points:
        if cv2.pointPolygonTest(black_contour, (point[0],point[1]), False) > 0:
            point[2] = True
    print(points)
            
    if points[0][2] is True and points[1][2] is True and points[2][2] is True and points and points[3][2] is True:
        print("test")
        random_shot = random.randint(1, directory_size_black)
        shot = cv2.imread(f'shootanalysisdataset/schotennobg/shotblack/{random_shot}.png', cv2.IMREAD_UNCHANGED)

    return shot
    
    
def generate_shots(amount):
    
    dir_path_white = r'C:\projects\python-cv\shootanalysisdataset\schotennobg\shotwhite'
    dir_path_black = r'C:\projects\python-cv\shootanalysisdataset\schotennobg\shotblack'
    
    directory_size_white = len([entry for entry in os.listdir(dir_path_white) if os.path.isfile(os.path.join(dir_path_white, entry))])
    directory_size_black = len([entry for entry in os.listdir(dir_path_black) if os.path.isfile(os.path.join(dir_path_black, entry))])
    
    background = shotgenerateimg.copy()
    counter = 0
    
    while counter < amount:
            
        max_width = shotgenerateimg.shape[1]
        max_height = shotgenerateimg.shape[0]
        
        startpoint_x = random.randint(0, max_width)
        startpoint_y = random.randint(0, max_height)
        endpoint_x = startpoint_x + 180
        endpoint_y = startpoint_y + 160
        
        points = [[startpoint_x, startpoint_y, False],
                  [startpoint_x, endpoint_y, False], 
                  [endpoint_x, startpoint_y, False],
                  [endpoint_x, endpoint_y, False]]
        
        shot_img_proposed = decide_shot_img(points, directory_size_white, directory_size_black)
                
        background = overlay_transparent(background, shot_img_proposed, startpoint_x, startpoint_y)
        
        #cv2.drawContours(background, [black_contour],  -1, (255,0,0), 2)

        counter += 1
    
    return background

background = generate_shots(5)
```
