```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img2 = cv2.imread('shootanalysisdataset/schotanalyse/20200923_130835.jpg',0)
resized_img = cv2.resize(img2, (round((img2.shape[1] / 4)), round((img2.shape[0] / 4))))

kernel = np.ones((10, 10), np.uint8)

img = cv2.erode(resized_img, kernel, iterations=1)
```

```python
img.shape
```

```python
plt.imshow(img,cmap='gray')
```

```python
contours,hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
contour_img = np.zeros(img.shape)
```

```python
type(contours)
```

```python
len(contours)
```

```python
contour_img
```

```python
image = cv2.drawContours(contour_img, contours, -1, (255,0,255), 1)
```

```python
type(hierarchy)
```

```python
hierarchy
```

```python
external_contours = np.zeros(image.shape)
```

```python
external_contours.shape
```

```python
plt.imshow(image)
```

```python
for i in range(len(contours)):
    
    #EXTERNAL contour
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours,contours,i,255,-1)
```

```python
plt.imshow(external_contours,cmap='gray')
```

```python
internal_contours = np.zeros(image.shape)

for i in range(len(contours)):
    
    #INTERNAL contour
    if hierarchy[0][i][3] == 4:
        cv2.drawContours(internal_contours,contours,i,255,-1)
```

```python
plt.imshow(internal_contours)
```
