# 1.1. K-Means

The first, and probably most famous, technique discussed was the k-means. 

 
To get started you need to select a amount of points to start equal of the K value. Lets presume you pick 2 points. Then for each of these points you calculate the distance of all other points in the dataset. Then the points get assigned a class based upon the lowest distance to either of the 2 k points. After all points have been assigned a class. The Mean for each class will be calculated. The same process will now happen again but instead of the two randomly picked points at the start. The Means for both classes are used.


We think there are two clusters, so lets first use that and set the k-means model using k=2. We'll use Elkan's algorithm, rather than full just to speed things along.

```python
# code to create and fit the k-means model (with elkan and full) 
#!pip install sklearn
from sklearn.cluster import KMeans

# enter code to select the two variables, make sure to also remove the missing values.
kmeans = KMeans(init='random', algorithm='elkan', n_clusters=2, random_state=0)
kmeans.fit(custX)
```

Great, now lets see the result...

```python
# code to create scatterplot with the classification
plt.scatter(custX["Finishing"],custX["HeadingAccuracy"], c=kmeans.labels_, cmap='rainbow');
```

This is ofcourse not really the result that we wanted. It is not all bad though. The reasons a part of the big blob is categorized wrong is because the Mean of the purple class had a lower distance. then the Mean of the red class.

Perhaps another initialization method works better. K-means++ is quite popular. 

K-means ++ is different because it has a smarter way of selecting the two Mean points. Instead of choosing two points at random it chooses on point. The distances to that one point are then calculated. The point with the biggest distance becomes the second point.

Lets see:

```python
# code to create and fit the k-means++ (elkan) and also show the scatterplot
kmeans = KMeans(init='k-means++',algorithm='elkan', n_clusters=2, random_state=0)
kmeans.fit(custX)

plt.scatter(custX["Finishing"],custX["HeadingAccuracy"], c=kmeans.labels_, cmap='rainbow');
```

```python
#!pip install yellowbrick
from yellowbrick.cluster import KElbowVisualizer
```

```python
# copy the elbowPlot function, and if you like you can adjust it.
def elbowPlot(data, model, kRange):

    # set the plotting stage for later
    fig, (ax1, ax2, ax3) = plt.subplots(1,3, constrained_layout=True,figsize=(15,5))

    # Using the Distortion measure:
    visualizer = KElbowVisualizer(model, k=kRange, metric='distortion', ax=ax1)
    chPlot=visualizer.fit(data)
    ax1.set_title('Distortion')

    # Using the Calinski-Harabasz measure
    visualizer = KElbowVisualizer(model, k=kRange, metric='calinski_harabasz', ax=ax2)
    chPlot=visualizer.fit(data)
    ax2.set_title('Calinski-Harabasz')

    # Using the Silhouette measure
    visualizer = KElbowVisualizer(model, k=kRange, metric='silhouette', ax=ax3)
    chPlot=visualizer.fit(data)
    ax3.set_title('Silhouette')

    # Show the results
    plt.show()

```

## 1.2. Hierarchical Clustering Analysis (HCA)

The second technique discussed was the Hierarchical Clustering Analysis (HCA). 

The method start of by making all points a cluster. Then it goes over all clusters and finds the two clusters closest together. These two clusters then become one. This step is then done over and over again. Until there is only one cluster.

One selection to be made is the so-called 'linkage'. The four basic ones are 'single', 'complete', 'mean', and 'ward'.

The single linkage looks at what point of a cluster is closest to another cluster. This is done for all clusters and the one with the lowest distance will then be merged.

The complete linkage works slightly different. It still looks for the smallest distance between two clusters but instead of looking at the closest point in the cluster it uses the furthest point.
    
The Mean is different in that it takes the average distance of all points of a cluster compared to all points of another cluster.
    
The ward method takes the variance of all points of a cluster. It does this for all clusters and then the lowest variance clusters are merged


In the Jupyter Notebook from the lecture there was a function that can compare these four options in one go. Lets use that here as well.

```python
# copy the compareLinkage function, and if you like you can adjust it.
from sklearn.cluster import AgglomerativeClustering

def compareLinkage(myData, k):
    hca, axs = plt.subplots(2,2, constrained_layout=True)

    hierClust = AgglomerativeClustering(linkage='single', n_clusters=k)
    hierClust.fit(myData)
    axs[0,0].scatter(myData[:,0],myData[:,1], c=hierClust.labels_, cmap='rainbow')
    axs[0,0].set_title('Single')

    hierClust = AgglomerativeClustering(linkage='average', n_clusters=k)
    hierClust.fit(myData)
    axs[0,1].scatter(myData[:,0],myData[:,1], c=hierClust.labels_, cmap='rainbow')
    axs[0,1].set_title('Average')

    hierClust = AgglomerativeClustering(linkage='complete', n_clusters=k)
    hierClust.fit(myData)
    axs[1,0].scatter(myData[:,0],myData[:,1], c=hierClust.labels_, cmap='rainbow')
    axs[1,0].set_title('Complete')

    hierClust = AgglomerativeClustering(linkage='ward', n_clusters=k)
    hierClust.fit(myData)
    axs[1,1].scatter(myData[:,0],myData[:,1], c=hierClust.labels_, cmap='rainbow')
    axs[1,1].set_title('Ward')

    plt.show()

```

Now lets use it on the Soccer data. We assumed there to be 2 clusters, so we'll stick with our k = 2 for now.

```python
# code to create the four HCA linkage models and scatterplot (use the function)
compareX = custX.to_numpy()
    
compareLinkage(compareX, 2)
```

## 1.3. DBSCAN

The third technique discussed was the Density Based Spatial Clustering of Applications with Noise (DBSCAN).

    
    
This technique works by choosing a random starting point. As parameters we need the epsilon and the min number of samples. The epsilon is the radius of the circle that will be used to check the amount of samples that fall within that circle. Lets say the number of samples is three points that means that within that circle 2 other points must be located to make it a core point. It then takes the closest point to the starting point and checks the same thing. It then takes the second closest to the starting point and so forth and so forth. If a point has less then the required number of samples that means that the point becomes a border or a outlier.

This method will require two input parameters. The epsilon and the min number of samples.

As the min number of samples I have chosen for 12

For the epsilon we can create a visual aid using the nearest neighbors.

```python
# code to show the nearest neighbors plot with different distances
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python

ns=12 # the minimum number of neighbors
nbrs = NearestNeighbors(n_neighbors=ns).fit(compareX)
distances, indices = nbrs.kneighbors(compareX)
distanceDec = sorted(distances[:,ns-1], reverse=True)
plt.plot(list(range(1,compareX.shape[0]+1)), distanceDec);
```

    
The elbow points seems to occur around 2 so lets use that.

Now that we have a min number of points for a neighborhood, and an epsilon we can run the DBSCAN.

```python
# code to create and fit the DBSCAN model.
dbscan = DBSCAN(eps=2,min_samples=12)
dbscan.fit(compareX)
```

Lets look at the results:

```python
# code to show the scatterplot with the clusters
plt.scatter(compareX[:,0],compareX[:,1], c=dbscan.labels_, cmap='rainbow');
```

## 1.4. HDBSCAN

The fourth technique was the Hierachical Density Based Spatial Clustering of Applications with Noise (HDBSCAN).


HDBSCAN works differently instead of checking if there are a certain amount of points within a distance. It calculates the distance for every point. This distance is calculated by checking the minimum amount of distance it takes to have the selected number of points. If it has done this for each point. Then every point is compared to every other point and it goes as follows. This distance will now be refered to as the radius. Point A had a radius and Point B had a radius for getting the selected number of points. Then the distance between Point A and B is calculated. If this is more then the radii of Point A and B that distance will be used. If not then the biggest radii Of either Point A or B will be used.

Once this has been done one by one the biggest distances will be removed. Everytime the distance is removed a check will be done to see if the point is still accessible from another point. If this is the case then it is removed. Otherwise it will be left in. This will be done until no point can be removed.

Based upon these distance a hierarchy will be build up.



One big advantage of this technique is that it only requires one input parameter: the min samples. We can use the same as from DBSCAN. So lets dive right in.

```python
# code to create and fit the HDBSCAN model.
import hdbscan

hdbscanModel = hdbscan.HDBSCAN(min_cluster_size=100, gen_min_span_tree=True)
hdbscanModel.fit(custX)
```

And the result....

```python
# code to show the scatterplot with the clusters
plt.scatter(custX.iloc[:,0],custX.iloc[:,1], c=hdbscanModel.labels_, cmap='rainbow');
```

This is probably the best one yet. It has generated the 2 clusters that you would expect when looking at the picture. The only downside is probably the outliers.

The HDBSCAN actually goes through a few steps: The minimum spanning tree => the dendrogram => the condensed tree.

In the Jupyter Notebook from the lecture, a small function was made to show these. 

```python
# code of the HDBSCANvisuals function, adjust if needed/wanted/improvements.
import seaborn as sns

# Remove missing values

hdbscanModel500 = hdbscan.HDBSCAN(min_cluster_size=100, gen_min_span_tree=True)
hdbscanModel500.fit(custX.head(500))

def HDBSCANvisuals():
    # set the plotting stage for later
    plt.figure(figsize=(10,3))

    plt.subplot(131)
    plt.gca().set_title('Min. Spanning Tree')
    hdbscanModel500.minimum_spanning_tree_.plot(edge_cmap='viridis', edge_alpha=0.6, node_size=80,edge_linewidth=2)

    #this gives a recursion error because of the size of the dataset
    plt.subplot(132)
    plt.gca().set_title('Dendrogram')
    hdbscanModel500.single_linkage_tree_.plot(cmap='viridis', colorbar=True)

    plt.subplot(133)
    plt.gca().set_title('Condensed Tree')
    hdbscanModel500.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())

    # Show the results
    plt.show()
```

The basic scatterplot can become a big blur, especially when you have a lot of datapoints.

*note: you can also use the clusterable_data.npy or another dataset, but it has to be one with a LOT of datapoints*

```python
# code to load dataset if not already loaded
# download the following https://www.kaggle.com/datasets/ulrikthygepedersen/diamonds
diamondsDF=pd.read_csv('diamonds.csv', sep = ',', header=0)
dfSel=diamondsDF[["price", "depth"]]
dfSel=dfSel.dropna()

from scipy import stats

#removes outliers for a cleaner result with the dataset
def drop_numerical_outliers(df, z_thresh=3):
    # Constrains will contain `True` or `False` depending on if it is a value below the threshold.
    constrains = df.select_dtypes(include=[np.number]) \
        .apply(lambda x: np.abs(stats.zscore(x)) < z_thresh) \
        .all(axis=1)
    # Drop (inplace) values set to be rejected
    df.drop(df.index[~constrains], inplace=True)
    
drop_numerical_outliers(dfSel)
```

Another method to deal with even larger datasets is a heatmap. An example is shown below.

![image heatmap](https://encrypted-tbn0.gstatic.com/images?q=tbn%3AANd9GcRvQZGu7SMVBCh6KIcF7toUqRpak2qWwQNkkhpZzv30hFbwYdDk&usqp=CAU)

This usually requires a Gaussian kernel, and is often slower than the previous methods.

So lets create a function for this as well.

```python
import matplotlib.cm as cm
from scipy.ndimage.filters import gaussian_filter

#creates the heatmap
def myplot(x, y, bins=300):
    heatmap, xedges, yedges = np.histogram2d(x, y, bins=bins)
    heatmap = gaussian_filter(heatmap, sigma=1)

    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
    return heatmap.T, extent
```

Here's an example of using the function:
