```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
shooting_card = cv2.imread('shootanalysisdataset/schotanalyse/20200923_130914.jpg')
```

```python
sep_blur = cv2.medianBlur(shooting_card,205)
```

```python
display(sep_blur)
```

```python
gray_sep_coins = cv2.cvtColor(sep_blur, cv2.COLOR_BGR2GRAY)
```

```python
display(gray_sep_coins)
```

```python
ret, sep_thresh = cv2.threshold(gray_sep_coins,130,220,cv2.THRESH_BINARY_INV)
```

```python
display(sep_thresh)
```

```python
image,contours,hierarchy = cv2.findContours(sep_thresh.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(shooting_card,contours,i,(255,0,0),10)
```
