```python
contours,hierarchy = cv2.findContours(sep_threshv2.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
contour_list = []
center_coordinates = np.zeros((0,3))


for x, contour in enumerate(contours):
    approx = cv2.approxPolyDP(contour,0.01*cv2.arcLength(contour,True),True)
    area = cv2.contourArea(contour)
    if ((len(approx) > 10) & (area > 1000) ):
        contour_list.append([x, contour])
        M = cv2.moments(contour)
        #GETS THE CENTER COORDINATES PLAN IS TO REMOVE OUTLIERS
        cx = int(M['m10']/M['m00'])
        cy = int(M['m01']/M['m00'])
        center_coordinates = np.vstack([center_coordinates, [cx, cy, x]])

x_median = np.median(center_coordinates[:,0])
y_median = np.median(center_coordinates[:,1])

print(x_median)
print(len(contour_list))

#removes contour from contour list if it is not close enough to the medium this is done to filter out contours
for coordinate in center_coordinates:

    if(((x_median * 1.1) < coordinate[0] or coordinate[0] < (x_median * 0.9)) or 
       ((y_median * 1.1) < coordinate[1] or coordinate[1] < (y_median * 0.9))):

        array_of_keys = np.array([inner_arr[0] for inner_arr in contour_list])
        indices = np.where(array_of_keys == int(coordinate[2]))
        
        mask = center_coordinates[:, 2] != coordinate[2]
        center_coordinates = center_coordinates[mask]
        
        contour_list.pop(indices[0][0])
        
        print(indices[0][0])
        print(coordinate[2])
        print(coordinate[0])
    

for contour in contour_list:
    M = cv2.moments(contour[1])
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])
    cv2.circle(resized_img, (cx, cy), 7, (255, 0, 255), -1)


#100
#120
#80
#200

contours_final = [subarr[1] for subarr in contour_list]

cv2.drawContours(img2, contours_final,  -1, (255,0,0), 2)


#OM DE CIRKEL NOG BETER TE DETECTEREN PROBEER HET VERSCHILLENDE BLUR LEVELS
#IDEE ALS NIET ALLE CIRKELS GEVONDEN WORDEN PROBEREN MET ANDERE BLUR


#double_contour = scale_contour(contour_list[0], 0.45)

#cv2.drawContours(img2, [double_contour],  -1, (255,0,0), 2)

plt.imshow(img2)
```

```python
img = cv2.imread('shootanalysisdataset/schotanalyse/20200923_130835.jpg')
```

```python
img = cv2.medianBlur(img,35)
```

```python
display(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
display(thresh)
```

```python
# NOISE REMOVAL
kernel = np.ones((3,3), np.uint8)
kernel
```

```python
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel,iterations=2)
```

```python
display(opening)
```

```python
sure_bg = cv2.dilate(opening,kernel,iterations=3)
```

```python
display(sure_bg)
```

```python
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
```

```python
display(dist_transform)
```

```python
ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
```

```python
display(sure_fg)
```

```python
sure_fg = np.uint8(sure_fg)
```

```python
unknown = cv2.subtract(sure_bg,sure_fg)
```

```python
display(unknown)
```

```python
ret , markers = cv2.connectedComponents(sure_fg)
```

```python
markers = markers + 1
```

```python
markers[unknown==255] = 0
```

```python
display(markers)
```
