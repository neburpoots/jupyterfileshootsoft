# 1. The Data
We are going to use the datafile diamonds.csv This contains data from diamonds

```python
# enter the code to load the data
# download the following https://www.kaggle.com/datasets/ulrikthygepedersen/diamonds
df=pd.read_csv('diamonds.csv', sep = ',', header=0)
```

Here's a quick look at the data:

```python

#ONLY TAKE THE TOP 3000 THIS DATASET HAS 60000 
#LINES AND WILL TAKE A LONG TIME OTHERWISE
df = df.head(3000)

# enter the code to see the first few rows of the data
df.head()
```

The dependent y will be the clarity of the diamond. We will predict the diamonds clarity with the two x values: price and carat.

The categorical dependent variable clarity has the following categories: 'b'I1'', 'b'IF'', 'b'SI1'', 'b'SI2'', 'b'VS1'', 'b'VS2'', 'b'VVS1'',
       'b'VVS2''

```python
# code that returns the different categories in the y variable.
#Create subset of data and remove missing values
dfSel=df[["carat","price", "clarity"]]
dfSel=dfSel.dropna()

dfSel["carat"]=pd.Categorical(dfSel["carat"])
dfSel["price"]=pd.Categorical(dfSel["price"])
dfSel["clarity"]=pd.Categorical(dfSel["clarity"])
dfSel.head()
```

```python

#get the numerical values as a numpy array
selNum = np.asarray(dfSel["clarity"].cat.codes)
selNum = np.dstack((selNum, np.asarray(dfSel["carat"].cat.codes)))
selNum = np.dstack((selNum, np.asarray(dfSel["price"].cat.codes)))
selNum = np.squeeze(selNum)

dfSel["clarity"].cat.categories
```

```python
np.asarray(dfSel["carat"].cat.codes)
```

We'll also need some training and testing data, so:

```python
# code to split data in training and testing
from sklearn.model_selection import train_test_split

# Now get X and Y:
X = selNum[:,1:3]
y = selNum[:,0:1]

# And convert them to a numpy array
X = np.asarray(X)
y = np.asarray(y)


#Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=109)
```

All set, lets try to predict this using our independent variables carat and price

## 2.3. K-Nearest Neighbors
The third technique is the K-Nearest Neighbors (KNN). 

KNN basically tries to categorize a value by looking what value/values is the closest. And then taking that category as well.

To use this we need to do some additional steps.

First we need to normalize our x variables:

```python
# use the StandardScaler to normalize the two x variables
# Import the standardscaler package
from sklearn.preprocessing import StandardScaler
# import the KNeighborsClassifier package
from sklearn.neighbors import KNeighborsClassifier


# set the scaler
scaler = StandardScaler()
scaler.fit(X_train)

# Convert the train and test X values, using the same scaler (so based on the X_train)
X_trainScaled = scaler.transform(X_train)
X_testScaled = scaler.transform(X_test)
```

Second we need to determine how many neighbors (k) we want. To do this we'll visualize the results using different values for k.

```python
# Code to create the graph with Error Rate vs. K-values.

error_rate=[]
for i in range(1,80):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_trainScaled, y_train.ravel())
    pred_i = knn.predict(X_testScaled)
    error_rate.append(np.mean(pred_i != y_test))

plt.figure(figsize=(10,6))
plt.plot(range(1,80),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

Takes a long time

```python

knnv2 = KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV
k_range = list(range(1,60))
leaf_size = list(range(40,60))
p=[1,2]

# I'm using gridsearch with only the k range or else it will take the rest of my life
# You could add leaf_size and p but it will take very long
param_grid = dict(n_neighbors=k_range)

# defining parameter range
grid = GridSearchCV(knnv2, param_grid, cv=10, scoring='accuracy', return_train_score=False,verbose=1)
  
# fitting the model for grid search
grid_search=grid.fit(X_trainScaled, y_train.ravel())

print(grid_search.best_params_)

accuracy = grid_search.best_score_ *100
print("Accuracy for our training dataset with tuning is : {:.2f}%".format(accuracy) )
```

Peter told us to look at the elbow point in the graph. According to gridsearch the best n_neighbors is 2

*Note:* +0.5 if you also use the GridSearch technique to decide on k.

Now that we know what we want k to be, we can create the model.

```python
# code to create the model with the selected k
knn = KNeighborsClassifier(n_neighbors=2)

knn.fit(X_trainScaled, y_train.ravel())
y_pred=knn.predict(X_testScaled)
```

Lets find out how good it is:

```python
# code to show its accuracy score AND confusion matrix.
print('accuracy score: ', metrics.accuracy_score(y_test, y_pred))

# The confusion matrix
plt.subplots(figsize=(20,15))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Blues', fmt='g')
plt.xlabel('predicted value')
plt.ylabel('actual value')
plt.xticks(np.arange(k)+1/k, labels, fontsize=18)
plt.yticks(np.arange(k)+1/k, labels, fontsize=18)
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```


The accuracy is quite good compared to the NB and SVM. You can clearly see that the values that most frequently appeared in the test data have been increased the most as well. This is logical when it categorizes on the closest ones.
