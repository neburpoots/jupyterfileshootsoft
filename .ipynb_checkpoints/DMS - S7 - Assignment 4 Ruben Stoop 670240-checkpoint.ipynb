# 1. The Item-Based Recommender

We are going to build an item-based recommender for Phones. In this recommender we could for example give people recommendations for phones based on a phone that they like.

To begin we'll need to load some data.

```python
# enter code to load your example data file(s)

#download https://www.kaggle.com/datasets/meirnizri/cellphones-recommendations?select=cellphones+data.csv

df = pd.read_csv('cellphones_data/cellphones users.csv', 
                 sep=',', 
                 encoding='latin-1', 
                 low_memory=False, 
                 )
df.head()
```

The dataset consists of three files<br><br>
cellphones data.csv<br>
   This contains the information about the phones themself. <br>
    
cellphones ratings.csv<br>
This contains information about the ratings of the different phones this also has a user_id and cellphone_id<br>
    
cellphones users.csv<br>
This contains all of the user information
    
<br>
We are going to get the ratings from the cellphones and combine these with the cellphones

```python
# Optional depending on your dataset, enter code to select appropriate columns, and merge the files.
ratings = pd.read_csv('cellphones_data/cellphones ratings.csv', 
                 sep=',', 
                 encoding='latin-1', 
                 low_memory=False,
                 error_bad_lines=False
                 )

ratings = ratings.rename(columns={"ï»¿cellphone_id": "cellphone_id"})



cellphones = pd.read_csv('cellphones_data/cellphones data.csv', 
                 sep=',', 
                 encoding='latin-1', 
                 low_memory=False,
                 error_bad_lines=False
                 )

cellphones = cellphones.rename(columns={"ï»¿cellphone_id": "cellphone_id"})
cellphones.head()
```

```python
#there we combine the ratings with the user
df2 = pd.merge(df, ratings, on='user_id')
df2.head()

#there we combine the ratings with the user
df3 = pd.merge(df2, cellphones, on='cellphone_id')
df3.head()
```

We need one row per user, and his/her scores for each Phone. We can do this by creating a pivot table.

```python
# code to create pivot table
phone_matrix = df3.pivot_table(index='user_id', columns='model', values='rating')
phone_matrix.head()
```

We also need to take the counts into consideration so lets keep those into their own dataframe, together with the average score:

```python
# code to make a dataframe with the unique objects, their average score and the number of scores
ratings = pd.DataFrame(df3.groupby('model')['rating'].mean())
ratings['nRatings'] = df3.groupby('model')['rating'].count()
ratings.sort_values(by=['nRatings']).head()
```

To decide on a cut-off point as minimum number of scores to be eligable we can plot the number of scores in a histogram:

```python
# code to generate the histogram
ratings['nRatings'].hist(bins=50, range=(10,50));
```

From the histogram we can see that that its not really necessary to remove any values. This dataset is not very big and all phones have got a good number of recommendations. If you really wanted too. You could use the code snipper below to remove the ones which have fewer then 25 ratings.

```python
ratings[ratings['nRatings']>=25].head()
```

Finally we can create the code to perform a recommendation. First we can select one particular phone

```python
# code to select one field
myPhone = phone_matrix['10 Pro']
```

And then to find the recommendation based on the correlations.

```python
# code to create the recommendation
myCorrs = phone_matrix.corrwith(myPhone).sort_values(ascending=False)

# and as a dataframe
corrDf = pd.DataFrame(myCorrs, columns=['Correlation'])
corrDf.dropna(inplace=True)
corrDf = corrDf.sort_values('Correlation', ascending=False)
corrDf.head()
```

The 10 pro has a perfect relation with the 10 pro. This is logical since it is the same phone. The Galaxy S22 Ultra is apparently the best recommendation if you like the 10 pro followed closely by the Xperia Pro. Then there is quite a gap to the next recommendation.

It might be nice to have one simple function that could do all the work for us. It should have as input parameters a dataframe, specific field, and minimum number of ratings, and then returns the recommendation.

Note that you should first select the fields that meet the minimum number of ratings, and then perform the correlation matrix, etc.

```python
# code for a python function
def recommendation(df, field, min_number_of_ratings, primary_key, rating_column, input_reco):
    
    # I've presumed this is the dataframe that is required as the input 
    #otherwise you would have to put the index and values as parameters in the function aswell.
    matrix = df.pivot_table(index=primary_key, columns=field, values=rating_column)
    
    # code to create pivot table
    myItem = matrix[input_reco]
    # and as a dataframe
    myCorrs = matrix.corrwith(myItem).sort_values(ascending=False)
    corrDf = pd.DataFrame(myCorrs, columns=['Correlation'])
    corrDf.dropna(inplace=True)
    corrDf = corrDf.sort_values('Correlation', ascending=False)
    
    #gets the number of ratings
    ratings = pd.DataFrame(df.groupby(field)[rating_column].mean())
    ratings['nRatings'] = df.groupby(field)[rating_column].count()
    ratings.sort_values(by=['nRatings']).head()
    
    corrDf = corrDf.merge(ratings['nRatings'], on=field)

    corrDf = corrDf[corrDf['nRatings']>=min_number_of_ratings]
    
    #return all the result without the input recommendation
    return corrDf[1:]

def new_recommendation(df, field, min_number_of_ratings):
    
    ratings = pd.DataFrame(df.groupby(field)['rating'].mean())
    ratings['nRatings'] = df.groupby(field)['rating'].count()
    corrDf = corrDf.merge(ratings['nRatings'], on='MovieName')
```

Lets test the function:
