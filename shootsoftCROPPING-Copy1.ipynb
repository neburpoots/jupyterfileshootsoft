```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
sep_coins = cv2.imread('shootanalysisdataset/schotanalyse/20200923_131036.jpg')
```

```python
display(sep_coins)
```

```python
# Median blur
# Grayscale
# Binary threshold
# Find Contours
```

```python
sep_blur = cv2.medianBlur(sep_coins,205)
```

```python
display(sep_blur)
```

```python
gray_sep_coins = cv2.cvtColor(sep_blur, cv2.COLOR_BGR2GRAY)
```

```python
display(gray_sep_coins)
```

```python
ret, sep_thresh = cv2.threshold(gray_sep_coins,160,220,cv2.THRESH_BINARY_INV)
```

```python
display(sep_thresh)
```

```python
contours,hierarchy = cv2.findContours(sep_thresh.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
contour_list = []
for contour in contours:
    approx = cv2.approxPolyDP(contour,0.01*cv2.arcLength(contour,True),True)
    area = cv2.contourArea(contour)
    if ((len(approx) > 15) & (area > 30) ):
        contour_list.append(contour)


cv2.drawContours(sep_coins, contour_list,  -1, (255,0,0), 2)
plt.imshow(sep_coins)

def scale_contour(cnt, scale):
    M = cv2.moments(cnt)
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])

    cnt_norm = cnt - [cx, cy]
    cnt_scaled = cnt_norm * scale
    cnt_scaled = cnt_scaled + [cx, cy]
    cnt_scaled = cnt_scaled.astype(np.int32)

    return cnt_scaled

double_contour = scale_contour(contour_list[0], 3.5)


plt.imshow(sep_coins)
```

```python
img = cv2.imread('shootanalysisdataset/schotanalyse/20200923_130835.jpg')
```

```python
img = cv2.medianBlur(img,35)
```

```python
display(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
display(thresh)
```

```python
# NOISE REMOVAL
kernel = np.ones((3,3), np.uint8)
kernel
```

```python
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel,iterations=2)
```

```python
display(opening)
```

```python
sure_bg = cv2.dilate(opening,kernel,iterations=3)
```

```python
display(sure_bg)
```

```python
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
```

```python
display(dist_transform)
```

```python
ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
```

```python
display(sure_fg)
```

```python
sure_fg = np.uint8(sure_fg)
```

```python
unknown = cv2.subtract(sure_bg,sure_fg)
```

```python
display(unknown)
```

```python
ret , markers = cv2.connectedComponents(sure_fg)
```

```python
markers = markers + 1
```

```python
markers[unknown==255] = 0
```

```python
display(markers)
```

```python
markers = cv2.watershed(img,markers)
```

```python
display(markers)
```

```python
image,contours,hierarchy = cv2.findContours(markers.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)

for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(sep_coins,contours,i,(255,0,0),10)
```

```python
import cv2
import numpy as np

# Load the image and convert it to grayscale
img = cv2.imread('target.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply a Gaussian blur to the grayscale image
gray_blur = cv2.GaussianBlur(gray, (5, 5), 0)

# Use the HoughCircles function to detect circles in the image
circles = cv2.HoughCircles(gray_blur, cv2.HOUGH_GRADIENT, dp=1, minDist=20, param1=50, param2=30, minRadius=10, maxRadius=100)

# Draw a circle around each detected circle
if circles is not None:
    circles = np.round(circles[0, :]).astype("int")
    for (x, y, r) in circles:
        cv2.circle(img, (x, y), r, (0, 255, 0), 2)

# Display the image with circles drawn around each detected circle
cv2.imshow("Circles", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
